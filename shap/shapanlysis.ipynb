Getting Started with the RDKit in Python

[![Powered by RDKit](https://img.shields.io/badge/Powered%20by-RDKit-3838ff.svg?logo=)](https://www.rdkit.org/)
    
    

```python
from rdkit import Chem
from ase.io import read as ase_reader
from ase.io import write as ase_write

# Read the CONTCAR file
structure = ase_reader('shap/CONTCAR', format='vasp')

atomic_numbers = structure.get_atomic_numbers()
```

```python
rdkit_mol = Chem.RWMol()
# 添加原子
for ase_atom in structure:
    # 获取原子序数
    atom_number = ase_atom.number
    # 获取原子符号
    symbol = Chem.GetPeriodicTable().GetElementSymbol(atom_number)
    # 添加原子到RDKit分子
    rdkit_mol.AddAtom(Chem.Atom(symbol))

# 添加化学键
for i, atom1 in enumerate(structure):
    for j, atom2 in enumerate(structure):
        if i != j:
            # 计算原子间距离
            distance = (atom1.position - atom2.position).norm()
            if distance < 2.0:  # 假设2.0 Å为化学键的阈值
                # 添加化学键到RDKit分子
                rdkit_mol.AddBond(i, j, Chem.BondType.SINGLE)

# 将RWMol对象转换为Mol对象
rdkit_mol = rdkit_mol.GetMol()

# 保存为Mol文件
with open('shap/atoms.mol', 'w') as f:
    Chem.MolToMolFile(rdkit_mol, f)
```

```python
from ase.io import read
from rdkit import Chem
import numpy as np


# 读取晶体结构
structure = read('shap/CONTCAR', format='vasp')

# 创建RDKit分子对象
rdkit_mol = Chem.RWMol()

# 添加原子
for ase_atom in structure:
    # 获取原子序数，并转换为Python的int类型
    atom_number = int(ase_atom.number)
    # 获取原子符号
    symbol = Chem.GetPeriodicTable().GetElementSymbol(atom_number)
    # 添加原子到RDKit分子
    rdkit_mol.AddAtom(Chem.Atom(symbol))

# 添加化学键
for i, atom1 in enumerate(structure):
    for j, atom2 in enumerate(structure):
        if i != j:
            # 计算原子间距离
            distance = np.linalg.norm(atom1.position - atom2.position)
            if distance < 2.0:  # 假设2.0 Å为化学键的阈值
                # 添加化学键到RDKit分子
                rdkit_mol.AddBond(i, j, Chem.BondType.SINGLE)

# 将RWMol对象转换为Mol对象
rdkit_mol = rdkit_mol.GetMol()

# 保存为Mol文件
with open('shap/atoms.mol', 'w') as f:
    Chem.MolToMolFile(rdkit_mol, f)
```
